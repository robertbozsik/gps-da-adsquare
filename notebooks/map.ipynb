```python
# imports
import pickle
import pandas as pd
import geopandas as gpd
import folium
```

# 3-4. Visualize the GPS signals of users and stores

```python
# open stores gdf
with open("../out_data/stores_gdf.pickle", "rb") as picke_file:
    stores_gdf = pickle.load(picke_file)
```

```python
# stores_gdf.head()
```

```python
# open final_df
final_df = pd.read_csv("../out_data/analysis.csv")
```

```python
# final_df.head()
```

```python
# create a map of Berlin
map_berlin = folium.Map(location=[52.5200,13.4050], zoom_start=11, tiles='cartodbpositron')

# display polygons of stores
stores_gdf.crs = "EPSG:4326" # Coordinate Reference System
folium.GeoJson(stores_gdf, 
               tooltip=folium.features.GeoJsonTooltip(
                   fields=['store_name'],
                   aliases=['store name:'])
              ).add_to(map_berlin);

# display latitudes and longitudes of device ids
for lat, lng, device_id in zip(final_df["lat"], final_df["lon"], final_df["device_id"]):
    tooltip = f"user: {device_id}"
    folium.CircleMarker(location=[lat, lng], 
                        radius=3, 
                        tooltip=tooltip, 
                        color="green",
                        fill=True, 
                        fill_color="green", 
                        fill_opacity=1
                       ).add_to(map_berlin)
```
