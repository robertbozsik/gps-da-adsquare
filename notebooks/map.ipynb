# 3-4. Visualize the GPS signals of users and stores

```python
stores_gdf = pd.read_pickle("../out_data/stores_gdf.pkl")
```

```python
# open final_df
users_in_stores = pd.read_csv("../out_data/users_in_stores.csv")
```

```python
users_in_stores.shape
```

```python
users_in_stores.head()
```

```python
# create a map of Berlin
map_berlin = folium.Map(location=[52.5200,13.4050], zoom_start=11, tiles='cartodbpositron')

# display polygons of stores
stores_gdf.crs = "EPSG:4326" # Coordinate Reference System
folium.GeoJson(stores_gdf, 
               tooltip=folium.features.GeoJsonTooltip(
                   fields=['store_name'],
                   aliases=['store name:'])
              ).add_to(map_berlin);

# display latitudes and longitudes of device ids
for lat, lng, device_id, store_name in zip(users_in_stores["lat"], users_in_stores["lon"], users_in_stores["device_id"], users_in_stores["store_name"]):
    tooltip = f"user: {device_id}"
    folium.CircleMarker(location=[lat, lng], 
                        radius=3, 
                        tooltip=tooltip, 
                        color="green",
                        fill=True, 
                        fill_color="green", 
                        fill_opacity=1
                       ).add_to(map_berlin)
```
